```python
import sys
import os
# add parent directory to sys.path
sys.path.append(os.path.dirname((os.getcwd())))
from output.circuit import inference
import json
```

```python
with open('../output/circuit.json') as f:
    circuit = json.load(f)

with open('y_test.json') as f:
    y_test = json.load(f)
```

```python
correct = 0
for i in range(10000):
    with open(f"X_test/{i}.json", "r") as f:
        input = json.load(f)
    out, _ = inference(input, circuit)
    correct += 1 if out[0] == y_test[i] else 0
    print(f"#{i}: {correct / (i + 1) * 100:.2f}%")
```
